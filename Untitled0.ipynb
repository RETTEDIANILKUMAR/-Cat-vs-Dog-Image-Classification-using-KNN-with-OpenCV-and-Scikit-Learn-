```python
# !pip install opencv-python
# !pip install PyWavelets

import cv2
import numpy as np
import pywt
import matplotlib.pyplot as plt

# Load grayscale image
image = cv2.imread("/content/drive/MyDrive/advancing image processing/WhatsApp Image 2025-07-22 at 14.48.35_8301ce80.jpg", cv2.IMREAD_GRAYSCALE)

# Show original image
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.show()

# Wavelet transform
wavelet = "db1"
coeffs2 = pywt.dwt2(image, wavelet)
LL, (LH, HL, HH) = coeffs2

titles = ["Approximation (LL)", "Horizontal (LH)", "Vertical (HL)", "Diagonal (HH)"]
images = [LL, LH, HL, HH]


plt.figure(figsize=(10, 6))
for i, (title, img) in enumerate(zip(titles, images)):
    plt.subplot(2, 2, i + 1)
    plt.imshow(img, cmap="gray")
    plt.title(title)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
